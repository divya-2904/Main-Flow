```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("C:\\Users\\TASK -- 3\\householdtask3.csv")
```

```python
data.head(10)
```

```python
#scatter plot with year against own
plt.scatter(data['year'],data['own'])
#adding title
plt.title("Scatter Plot")
#setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')
#show the scatter plot
plt.show()
```

```python
#line chart with year and own
plt.plot(data['year'])
plt.plot(data['own'])
#adding title
plt.title("line chart")
#setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')
#show the scatter plot
plt.show()
```

```python
#bar chart or bar plot
plt.bar(data['year'],data['own'])
#adding title
plt.title("Bar Chart")
#setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')
#show the scatter plot
plt.show()
```

```python
#histogram
plt.hist(data['income'])
plt.title("Histogram")
plt.show()
```
